```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 1. BAR PLOT
# Memvisualisasikan data dengan barplot
# dengan 'x' adalah sex, dan 'y' adlah total_bill
# dan datanya adalah tips
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# 2. COUNTER PLOT
# Memvisualisasikan data dengan countplot
# dengan 'x' adalah sex, dan datanya adalah tips
sns.countplot(x='sex', data=tips)
```

```python
# 3. BOX PLOT
# Memvisualisasikan data dengan boxplot
# dengan 'x' adalah sex, dan 'y' adlah total_bill, datanya adalah tips
# pallete adalah warna masing2 plot
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
# Memvisualisasikan data dengan boxplot secara orientasinyaHorizontal
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
# Memvisualisasikan data dengan boxplot
# hue akan menampilkan info dari warna data
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
```

```python
# 4. VIOLIN
# Memvisualisasikan data dengan violin plot
sns.violinplot(x='day', y='total_bill', data=tips, pallete='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', pallete='Setl')
```

```python
# Split true menggabungkan 2 tampilan yang sebelumnya terpisah
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True, pallete='Setl')
```

```python
# 5. STRIP PLOT
# Memvisualisasikan data dengan stripplot
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# Apabila jitter bernilai false maka akan membuat tampilannya menjasi 1 baris
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
# dodge = True, sama halnya spt SPLIT
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
# SWARM PLOT
# Memvisualisasikan data menggunakan swarmplot
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue='sex', data=tips, dodge=True)
```

```python
# Kombinasi antara violinplot dengan swarmplot
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=3)
```

```python
# 6. CAT PLOT
# Memvisualisasikan data dengan catplot
# Kind adalah jenis tampilannya
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar')
```
