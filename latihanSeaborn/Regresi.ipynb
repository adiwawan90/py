```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
tips.head()
```

```python
# LMPLOT
# Memvisualisasikan data dengan lmplot
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
```

```python
# Mengubah symbol hue dengan marker
# Scatter_kws adalah besar marker
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o','v'], scatter_kws={'s':100})
```

```python
# Memisahkan data
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', row='sex', data=tips, col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',  hue='sex', palette='coolwarm')
```

```python
# aspect adalah lebar grid
# height adalah tinggi grid
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', palette='coolwarm', aspect=0.6, height=5 )
```
