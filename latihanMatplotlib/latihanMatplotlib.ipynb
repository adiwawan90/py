```python
# LATIHAN MATPLOTLIB
import matplotlib.pyplot as plt
```

```python
import numpy as np
# Membuat data Array dimulai dr 1-5 dengan jumlah index ada 11
x = np.linspace(1, 5, 11)
# dari data array diatas diberi pangkat 2
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r') # r merupakan warna, red, b=blue
plt.xlabel('X Axis') #label untuk sumbu X
plt.ylabel('Y Axis') #label untuk sumbu Y
plt.title('Data Contoh') #label untuk jdul diagram
plt.show() #untuk memunculkan diagram
```

```python
# Membuat banyak diagram / plot dalam 1 canvas
plt.subplot(1,2,1) # *1 adalah row, *2 adlh col, *1 adlh plot
plt.plot(x, y, 'r--') # r-- adlh garis berwarna merah putus -putus
plt.subplot(1,2,2) # *1 adalah row, *2 adlh col, *2 adlh plot
plt.plot(x, y, 'b*-') # r-- adlh garis berwarna biru campur bintang dan garis
```

```python
# Membuat canvas dan ukuran plot
# Membuat canvas kosong
fig = plt.figure()
# Mengatur ukuran cnvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# 0.1 = jarak dari kiri
# 0.1 = jarak dari bawah
# 0.8 = lebar
# 0.8 = tinggi
axes.plot(x, y, 'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
# PLOTdi dalam PLOT
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Plot Utama
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #Plot didalam

# Pengaturan plot Utama
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Utama')
axes1.set_ylabel('Y Utama')
axes1.set_title('Plot Utama')

# Pengaturan Plot
axes2.plot(x, y, 'r')
axes2.set_xlabel('X Dalam')
axes2.set_ylabel('Y Dalam')
axes2.set_title('Plot Dalam')
```

```python
# Mengatur Canvas
# figsize adalah pengaturan ukuran terhadap canvas yang akan dibuat
# dpi adlah bentukan dari suatu ukuran hasil cetak / visual semakin tinggi maka semakin besar gambarnya
fig = plt.figure(figsize=(8,4), dpi=(100))
```

```python
fig, axes = plt.subplots(figsize=(8,4))
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
# Save atau export figure
fig, axes = plt.subplots(figsize=(8,4))
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

# save atau export figure menjadi format png, jpg, eps, svg, pgf, & pdf
fig.savefig("contoh.png", dpi=200)
```

```python
# menambahkan informasi pada plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='Data xpangkat 2')
ax.plot(x, x**3, label='Data x pangkat 3')

# menambahkan info ke dalam figure
ax.legend(loc=0)
# 'loc' posisi dimana kita akan menaruhnya, jika
# 0: default kiri atas
# 1: kanan atas
# 2: kiri atas
# 3: kiri bawah
# 4: kanan bawah
```

```python
# Style Plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color="red", linewidth=0.25) #linewidth mengatur ketebalan garis
ax.plot(x, x+2, color="red", linewidth=0.50) #linewidth mengatur ketebalan garis
ax.plot(x, x+3, color="red", linewidth=1.00) #linewidth mengatur ketebalan garis
ax.plot(x, x+4, color="red", linewidth=2.00) #linewidth mengatur ketebalan garis
```

```python
# Style Plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color="green", linewidth=3, linestyle='-') #linewidth bisa disingkat 'lw'
ax.plot(x, x+2, color="green", lw=3, ls='-.') #linestyle / ls mengatur style
ax.plot(x, x+3, color="green", lw=3, ls=':') 
```

```python
# Style Plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5,5,15,5]) #format ([pjg ls, jarak ls, pjg garis, jrak garis])
```

```python
# Style Plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# memberikan variasi simbols pada garis atau plot
ax.plot(x, x+1, color="red", lw=3, ls='-', marker='+') 
ax.plot(x, x+2, color="red", lw=3, ls='--', marker='o') 
ax.plot(x, x+3, color="red", lw=3, ls='-', marker='s') 
ax.plot(x, x+4, color="red", lw=3, ls='--', marker='1') 
```

```python
# Style Plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# markersize= besar marker
# markerfacecolor= warna marker
# markeredgecolor= besar border marker
# markeredgecolor= warna border marker
ax.plot(x, x+1, color="purple", lw=1, ls='-', marker='o', markersize=10) 
ax.plot(x, x+2, color="purple", lw=1, ls='-', marker='o', markersize=4) 
ax.plot(x, x+3, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor= 'red') 
ax.plot(x, x+4, color="purple", lw=1, ls='-', marker='s', markersize=10, markerfacecolor= 'yellow', markeredgewidth=3, markeredgecolor= 'pink' ) 
```
